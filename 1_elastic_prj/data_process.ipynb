```python
# 1. 파일을 전체적으로 문자열로 읽기
with open('dummy_data.json', 'r', encoding='utf-8') as f: #utf-8로 읽기
    data = f.read()
    print(data[0:100])

# 2. JSON 객체를 구분하는 문자열을 찾아서 줄 바꿈 문자로 교체
data = data.replace('},{', '},\n{')

# 3. 결과 문자열을 다시 파일에 쓰기
with open('dummy_data2.json', 'w', encoding='utf-8') as f:
    f.write(data)
```
